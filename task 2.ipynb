1.unique words in sentence

```python
# Step 1: Input sentence
sentence = input("Enter sentence: ")
# Step 2: Split into words
words = sentence.split()
# Step 3: Convert to set (unique words)
unique_words = set(words)
# Step 4: Count unique words
count = len(unique_words)
# Step 5: Display results
print("Unique words count:", count)
print("Unique words:", unique_words)
```

2.Highest Salary from Employee Data


```python
# Step 1: Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Step 2: Find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]
# Step 3: Display result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

3.Find Maximum and Minimum Values


```python
# Step 1: Input list
numbers = [45, 22, 89, 10, 66]
# Step 2: Initialize max and min
maximum = minimum = numbers[0]
# Step 3: Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
# Step 4: Display results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

4. Count Products Above a Price Threshold


```python
# Step 1: Input list
prices = [450, 1200, 899, 1500, 300]
# Step 2: Initialize counter
count = 0
# Step 3: Loop through prices
for price in prices:
    if price > 1000:
        count += 1
# Step 4: Display result
print("Products above 1000:", count)

```

5. Calculate Attendance Percentage


```python
# Step 1: Input list
attendance = ["P", "P", "A", "P", "P"]
# Step 2: Count presents
present_days = attendance.count("P")
# Step 3: Calculate percentage
total_days = len(attendance)
percentage = (present_days / total_days) * 100
# Step 4: Display result
print("Attendance Percentage:", percentage)
```

7.Count Character Frequency


```python
# Step 1: Input string
text = "python"
# Step 2: Initialize dictionary
char_frequency = {}
# Step 3: Loop through characters
for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1
# Step 4: Display result
print(char_frequency)
```

9.Check if a Key Exists in a Dictionary


```python
# Step 1: Input dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Step 2: Key to check
key_to_check = "Ravi"
# Step 3: Check existence
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

 10.Calculate Average Marks


```python
# Step 1: Input list
marks = [80, 70, 65, 75]
# Step 2: Calculate sum
total = sum(marks)
# Step 3: Count number of marks
count = len(marks)
# Step 4: Calculate average
average = total / count
# Step 5: Display result
print("Average Marks:", average)
```
