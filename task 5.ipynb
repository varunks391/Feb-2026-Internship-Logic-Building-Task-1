```python
# 1. Social Media – Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    # Step 1: Calculate total likes using a loop
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    # Step 2: Determine post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    # Step 3: Print results
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")
# Example usage
likes_data = [120, 300, 250, 380]  # Engagement data from different sources
analyze_post_engagement(likes_data)
```

```python
# 3.  Agriculture – Rainfall Adequacy Checker 

def rainfall_adequacy_checker(rainfall_data, required_level):
    # Step 1: Calculate total rainfall using a loop
    total_rainfall = 0
    for rainfall in rainfall_data:
        total_rainfall += rainfall
    # Step 2: Compute average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)
    # Step 3: Check adequacy
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    # Step 4: Print results
    print(f"Average Rainfall: {average_rainfall:.0f}")
    print(f"Rainfall Status: {status}")
# Example usage
rainfall_data = [80, 65, 70, 73]   # Seasonal rainfall data
required_level = 70                # Required adequacy threshold
rainfall_adequacy_checker(rainfall_data, required_level)
```

```python
# 4. Social Media – Duplicate Account Detection

def detect_duplicate_accounts(usernames):
    # Step 1: Use a set to track unique usernames
    unique_usernames = set()
    duplicate_found = False
    # Step 2: Loop through usernames
    for user in usernames:
        if user in unique_usernames:
            duplicate_found = True
            break
        else:
            unique_usernames.add(user)
    # Step 3: Print results
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
# Example usage
usernames_list = ["alice", "bob", "charlie", "alice"]  # Sample data
detect_duplicate_accounts(usernames_list)
```

```python
# 6. Agriculture – Premium Crop Price Filter 

def premium_crop_filter(crop_prices, threshold=2000):
    # Step 1: Filter premium crops using list comprehension
    premium_crops = [price for price in crop_prices if price > threshold]
    # Step 2: Print results
    print(f"Premium Crops: {premium_crops}")
# Example usage
crop_prices = [1500, 2500, 1800, 3200, 2000]  # Sample crop prices
premium_crop_filter(crop_prices)
```

```python
# 9. E-Learning – Student Attendance Eligibility System

def attendance_eligibility(attendance_records, total_classes):
    # Step 1: Count attended classes using a loop
    attended_classes = 0
    for attended in attendance_records:
        if attended:   # True means attended, False means absent
            attended_classes += 1
    # Step 2: Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    # Step 3: Check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    # Step 4: Print results
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {status}")
# Example usage
attendance_records = [True, True, False, True, True]  # 5 classes, attended 4
total_classes = 5
attendance_eligibility(attendance_records, total_classes)
```

```python
# 10. Smart Electricity Bill Analyzer 

def electricity_bill(units):
    # Step 1: Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    # Step 2: Classify usage level
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    # Step 3: Return results
    return bill, status
# Example usage
units_consumed = 250
bill, usage_status = electricity_bill(units_consumed)
print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{bill}")
print(f"Usage Status: {usage_status}")
```
