```python
# 2.Search Query Keyword Analysis
import string
# Input query
query = "Buy mobile phone buy phone online"
# Step 1: Convert to lowercase
query = query.lower()
# Step 2: Remove punctuation
query = query.translate(str.maketrans("", "", string.punctuation))
# Step 3: Split into words
words = query.split()
# Step 4: Count frequency of each keyword
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
# Step 5: Filter keywords searched more than once
result = {word: count for word, count in frequency.items() if count > 1}
# Step 6: Display result
print(result)
```

```python
# 1. Employee performance dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Step 1: Find the highest score
max_score = max(employees.values())
# Step 2: Identify all employees with the highest score
top_performers = [name for name, score in employees.items() if score == max_score]
# Step 3: Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

```python
# 3.Sensor Data Validation
# Input sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# Step 1: Identify valid (even) readings
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]
# Step 2: Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# 4. Email Domain Usage Analysis
# Input email list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
# Step 1: Extract domains
domains = [email.split("@")[1] for email in emails]
# Step 2: Count frequency of each domain
domain_count = {}
for domain in domains:
    domain_count[domain] = domain_count.get(domain, 0) + 1
# Step 3: Calculate percentage usage
total = len(emails)
domain_percentage = {domain: (count / total) * 100 for domain, count in domain_count.items()}
# Step 4: Display results
for domain, percentage in domain_percentage.items():
    print(f"{domain}: {int(percentage)}%")
```

```python
# 5.Sales Spike Detection
# Input sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]
# Step 1: Calculate average daily sales
average_sales = sum(sales) / len(sales)
# Step 2: Detect spikes (sales > 30% above average)
spikes = [(day + 1, value) for day, value in enumerate(sales) if value > average_sales * 1.3]
# Step 3: Display results
print("Sales Spikes Detected:")
for day, value in spikes:
    print(f"Day {day}: {value}")
```

```python
# 6. Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Count occurrences
duplicates = {}
for uid in user_ids:
    duplicates[uid] = duplicates.get(uid, 0) + 1
print("Problem 6 Output:")
for uid, count in duplicates.items():
    if count > 1:
        print(f"{uid} → {count} times")
print("-" * 50)
```
