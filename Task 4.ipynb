1. mobile recharge validation system(1)

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans are: 199, 299, 399, 599")
        return False
    else:
        print("Recharge successful for ₹", amount)
        return True
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(amount):
            break

        print("Please try again.\n")

    except ValueError:
        print("Invalid input! Please enter a valid number.\n")
```

2. inventory reorder alert system

```python
def check_inventory(products):
    print("Inventory Status:\n")

    for product, stock in products.items():
        if stock < 15:
            print(product, "-> Reorder Alert (Stock:", stock, ")")
        else:
            print(product, "-> Stock OK (Stock:", stock, ")")
inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 25,
    "Oil": 8,
    "Salt": 18
}
check_inventory(inventory)
```

3.student result processing system

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"

student_marks = [60, 45, 70, 55, 80]

avg, result = process_result(student_marks)

print("Student Marks:", student_marks)
print("Average Marks:", avg)
print("Result:", result)
```

4. cab fare estimator with retry option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare
while True:
    try:
        distance = float(input("Enter distance in km: "))

        if distance <= 0:
            print("Distance must be greater than 0.\n")
            continue

        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)

        print("Total Cab Fare: ₹", round(total_fare, 2))
        break

    except ValueError:
        print("Invalid input! Please enter valid numbers.\n")
```

5. employee attendance eligiblity

```python
def check_eligibility(attendance):
    present_days = 0
    for day in attendance:
        if day.upper() == "P":
            present_days += 1

    total_days = len(attendance)
    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"

attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]
percentage, status = check_eligibility(attendance_list)

print("Attendance List:", attendance_list)
print("Attendance Percentage:", round(percentage, 2), "%")
print("Eligibility Status:", status)
```

6. password strength checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    if len(password) < 8:
        return "Weak Password (Minimum length should be 8)"
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include at least one digit and one special character @#$)"
user_password = "Pass@123"
result = check_password(user_password)

print("Password:", user_password)
print("Strength:", result)
```
