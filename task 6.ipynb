```python
# 1.Smart Parking Lot Management System
def smart_parking_system(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0
    # Process each log
    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1
        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked
    # Display results
    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)
    # Capacity check
    if current_parked > capacity:
        print("ALERT: Parking capacity exceeded!")
    else:
        print("Parking Status: Available")
# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

```python
# 2.Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    # Base time calculation
    base_time = distance * 5
    # Traffic adjustment
    if traffic == "Low":
        base_time += 5
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "High":
        base_time += 15
    # Weather adjustment
    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20
    print("Estimated Delivery Time:", base_time, "minutes")
# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"
estimate_delivery_time(distance, traffic, weather)
```

```python
# 3. Movie Theatre Seat Occupancy Analyzer
def seat_occupancy_analyzer(total_seats, booked_seats):
    # Count booked seats
    booked_count = len(booked_seats)
    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    print("Occupancy:", int(occupancy), "%")
    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional shows.")
    else:
        print("Show Status: Seats Available")
# Sample Input
total_seats = 200
booked_seats = [1] * 150  # 150 booked seats
seat_occupancy_analyzer(total_seats, booked_seats)
```

```python
#  4.Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total = 0
    # Calculate total using loop
    for reading in cpu_readings:
        total += reading
    # Calculate average
    average = total / len(cpu_readings)
    print("Average CPU Load:", int(average), "%")
    # Classification rules
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
# Sample Input
cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```

```python
# 5.mart Classroom Resource Usage Monitor
def resource_usage_monitor(resource_usage):
    overused_resources = []
    # Check each resource using dictionary
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused_resources.append(resource)
    # Display overused resources
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
resource_usage_monitor(resource_usage)
```

```python
# 6.
def registration_controller(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0
    # Process registrations one by one
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    # Registration status
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")
# Sample Input
capacity = 100
registrations = 105
registration_controller(capacity, registrations)
```
